```python
import pandas as pd

# Load data from CSV
file_path = '/Users/hpourmand/Desktop/Data Analyst Sample/checkouts.csv'
df = pd.read_csv(file_path)

# Display basic information about the DataFrame
df.info()

# Check for missing values and print the count for each column
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Drop rows with missing values (uncomment if needed)
df = df.dropna()   

# Filter rows based on date range
start_date = '2017-01-01'
end_date = '2020-12-24'
df = df[(df['date_checkout'] >= start_date) & (df['date_checkout'] <= end_date)]
df = df[(df['date_returned'] >= start_date) & (df['date_returned'] <= end_date)]


# Add filter to ensure 'date_checkout' is before 'date_returned'
df = df[df['date_checkout'] < df['date_returned']]


# Clean up columns by stripping unwanted characters and whitespace
columns_to_strip = ['id', 'patron_id', 'library_id', 'date_checkout', 'date_returned']
for column in columns_to_strip:
    if column in df.columns:
        df[column] = df[column].astype(str).str.strip().str.strip('%').str.replace('/', '-')
        df[column] = df[column].astype(str).str.replace('|', '-').str.replace(' ', '-')



# Convert 'date_checkout' and 'date_returned' to datetime format and then format to %Y-%m-%d
date_columns = ['date_checkout', 'date_returned']
for column in date_columns:
    if column in df.columns:
        df[column] = pd.to_datetime(df[column], errors='coerce').dt.strftime('%Y-%m-%d')
        df[column] = df[column].fillna('')


# Save cleaned DataFrame to CSV
output_path = '/Users/hpourmand/Desktop/Data Analyst Sample/cleaned_checkouts.csv'
df.to_csv(output_path, index=False)

print(df)
```
