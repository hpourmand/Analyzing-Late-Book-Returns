```python
import pandas as pd

# Load data from CSV
file_path = '/Users/hpourmand/Desktop/Data Analyst Sample/customers.csv'
df = pd.read_csv(file_path)




# Display basic information about the DataFrame
df.info()

 # Drop duplicates
df = df.drop_duplicates()

# Check for missing values and print the count for each column
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)


# Filling missing values
df['city'] = df['city'].fillna('Unknown')
df['state'] = df['state'].fillna('Unknown')
#df['gender'] = df['gender'].fillna('Not Provided')
df['education'] = df['education'].fillna('Others')
df['occupation'] = df['occupation'].fillna('Not Provided')

# Fill missing values in 'birth_date' with the mode
birth_date_mode = df['birth_date'].mode()[0]
df['birth_date'] = df['birth_date'].fillna(birth_date_mode)




# Filter dates
Latest_date = '2020-12-29'
Earliest_date = '1900-01-01'
df = df[(df['birth_date'] <= Latest_date) & (df['birth_date'] >= Earliest_date)]



# Drop rows with other missing values
df = df.dropna()

# Define a function to standardize column values
def standardize_column(df, column_name, replacements):
    for old_value, new_value in replacements.items():
        df[column_name] = df[column_name].str.replace(old_value, new_value, case=False, regex=True)
    df[column_name] = df[column_name].str.strip()
    df[column_name] = df[column_name].str.replace(r'\s+', ' ', regex=True)  


# Standardize values in columns
standardize_column(df, 'gender', {'female': 'FEMALE', 'male': 'MALE'})
standardize_column(df, 'city', {'PORTLAND': 'Portland'})
standardize_column(df, 'state', {'OREGON': 'Oregon', 'washington': 'Washington'})
standardize_column(df, 'education', {
    'COLLEGE': 'College', 'SCHOOL': 'School', 'HIGH': 'High',
    'GRADUATE': 'Graduate', 'graduate': 'Graduate', 'DEGREE': 'Degree',
    'OTHERS': 'Others'
})
standardize_column(df, 'occupation', {
    'OTHERS': 'Others', 'TECH': 'Tech', 'tech': 'Tech', 'COLLAR': 'Collar',
    'SUPPORT': 'Support', 'ADMIN': 'Admin', 'BLUE': 'Blue',
    'FINANCE': 'Finance', 'EDUCATION': 'Education', 'education': 'Education',
    'sales': 'Sales', 'SALES': 'Sales'
})

# Clean up columns by stripping unwanted characters and whitespace
columns_to_strip = ['zipcode', 'name', 'gender', 'city', 'street_address', 'state', 'birth_date', 'education', 'occupation']
for column in columns_to_strip:
    df[column] = df[column].str.strip().str.strip('_#-')


# Convert 'zipcode' column to integer
df['zipcode'] = df['zipcode'].astype(str).str.split('.').str[0]  


# Save cleaned DataFrame to CSV
output_path = '/Users/hpourmand/Desktop/Data Analyst Sample/cleaned_customers.csv'
df.to_csv(output_path, index=False)

print(df)
```
